```python
import pandas as pd
```

```python
df = pd.read_csv('ign_reviews_final.csv')
df
```

```python
#show range of dates
print(df['date'].min())
print(df['date'].max())
```

```python
# plot a histogram of scores
df['score'].plot(kind='hist', bins=20, title='IGN Review Scores')
```

```python
# plot a histogram of scores
df['score'].plot(kind='hist', bins=35, title='IGN Review Scores')
```

```python
#mean score
print(df['score'].mean())
```

```python
#median score
print(df['score'].median())
```

```python
#group by author and get the median score, sorted by median
df.groupby('author')['score'].median().sort_values()
```

```python
df[df['author'] == 'Jaz Rignall']
```

```python
#group by author and get the number of articles, sorted by number of articles
df.groupby('author')['score'].count().sort_values()
```

```python
#reviews for every pokemon game
df_pokemon = df[df['title'].str.contains('Pokemon')]
df_pokemon
```

```python
#sort by date
df_pokemon.sort_values('date')
```

```python
# plot a bar chart of scores, with the x-axis being date
df_pokemon.sort_values('date').plot(x='date', y='score', kind='line', title='Pokemon Review Scores')
```
