## comparing ign with metacritic

```python
from fuzzywuzzy import process
import pandas as pd

def find_similar_games(ign_titles, metacritic_titles, threshold=90):
    similar_games = []
    for title in ign_titles:
        match = process.extractOne(title, metacritic_titles, score_cutoff=threshold)
        if match:
            similar_games.append((title, match[0]))
    return similar_games

# Load your datasets here
metacritic_scores = pd.read_csv('metacritic-scores.csv')
ign_reviews = pd.read_csv('ign_reviews_final.csv')

# Extract game names
metacritic_game_names = set(metacritic_scores['name'].str.strip().unique())
ign_game_titles = set(ign_reviews['title'].str.split(':').str[0].str.strip().unique())

# Convert to lists for processing
ign_titles_list = list(ign_game_titles)
metacritic_titles_list = list(metacritic_game_names)

# Find similar but not exactly matching titles
similar_games = find_similar_games(ign_titles_list, metacritic_titles_list, threshold=90)

# Create a DataFrame from the similar games list
similar_games_df = pd.DataFrame(similar_games, columns=['IGN_Title', 'Metacritic_Title'])

# Export the DataFrame to a CSV file
similar_games_df.to_csv('similar_games.csv', index=False)

print("Exported similar games to 'similar_games.csv'.")
```

```python
# Load the datasets
metacritic_scores = pd.read_csv('metacritic-scores.csv')
ign_reviews = pd.read_csv('ign_reviews_final.csv')

# Find overlaps where Metacritic names appear within IGN titles
overlaps_within_ign_titles = [name for name in metacritic_scores['name'].unique() if any(name in title for title in ign_reviews['title'])]

# Filter the original datasets for the overlaps
metacritic_filtered = metacritic_scores[metacritic_scores['name'].isin(overlaps_within_ign_titles)]
ign_filtered = ign_reviews[ign_reviews['title'].apply(lambda title: any(name in title for name in overlaps_within_ign_titles))]

# Merge the two filtered datasets on the condition that the Metacritic game name appears within the IGN title
merged_overlaps = pd.merge(metacritic_filtered, ign_filtered, how='inner', left_on='name', right_on='title')

# Selecting a subset of columns for the final DataFrame
selected_columns = ['name', 'platform', 'release_date', 'metascore', 'user_score', 'title', 'score', 'url', 'datetime', 'author', 'date']
merged_overlaps_selected = merged_overlaps[selected_columns]

# Export the merged DataFrame to a CSV file
output_file_path = 'merged_overlaps_games.csv'
merged_overlaps_selected.to_csv(output_file_path, index=False)

print(f"File saved as {output_file_path}.")
```
