```python
def clean_and_export_csv(input_file_path, output_file_path):
    """
    Reads a CSV file, removes duplicate titles, and exports the cleaned data to a new CSV file.

    Parameters:
    - input_file_path: str, the path to the input CSV file.
    - output_file_path: str, the path to the output CSV file where the cleaned data will be saved.

    Returns:
    None
    """
    # Read the input CSV file
    df = pd.read_csv(input_file_path)

    # Drop duplicates based on the 'title' column, keeping the first occurrence
    df_cleaned = df.drop_duplicates(subset='title', keep='first')

    # Export the cleaned DataFrame to a CSV file without the index
    df_cleaned.to_csv(output_file_path, index=False)
```

```python
clean_and_export_csv('data/ign_ps4.csv', 'data/cleaned/ign_ps4_cleaned.csv')
clean_and_export_csv('data/ign_ps3.csv', 'data/cleaned/ign_ps3_cleaned.csv')
clean_and_export_csv('data/ign_playstation2.csv', 'data/cleaned/ign_ps2_cleaned.csv')
clean_and_export_csv('data/ign_playstation.csv', 'data/cleaned/ign_ps1_cleaned.csv')
clean_and_export_csv('data/ign_pc.csv', 'data/cleaned/ign_pc_cleaned.csv')
clean_and_export_csv('data/ign_n64.csv', 'data/cleaned/ign_n64_cleaned.csv')
clean_and_export_csv('data/ign_ngage.csv', 'data/cleaned/ign_ngage_cleaned.csv')
clean_and_export_csv('data/ign_mac.csv', 'data/cleaned/ign_mac_cleaned.csv')
clean_and_export_csv('data/ign_iphone.csv', 'data/cleaned/ign_iphone_cleaned.csv')
clean_and_export_csv('data/ign_ipad.csv', 'data/cleaned/ign_ipad_cleaned.csv')
clean_and_export_csv('data/ign_gba.csv', 'data/cleaned/ign_gba_cleaned.csv')
clean_and_export_csv('data/ign_gb.csv', 'data/cleaned/ign_gb_cleaned.csv')
clean_and_export_csv('data/ign_gamecube.csv', 'data/cleaned/ign_gamecube_cleaned.csv')
clean_and_export_csv('data/ign_ds.csv', 'data/cleaned/ign_ds_cleaned.csv')
clean_and_export_csv('data/ign_dreamcast.csv', 'data/cleaned/ign_dreamcast_cleaned.csv')
clean_and_export_csv('data/ign_android.csv', 'data/cleaned/ign_android_cleaned.csv')
clean_and_export_csv('data/ign_3ds.csv', 'data/cleaned/ign_3ds_cleaned.csv')
```

# Let's merge all the cleaned csvs

```python
#merge all the cleaned csvs into one

# Create a list of file paths
file_paths = [
    'data/cleaned/ign_ps4_cleaned.csv',
    'data/cleaned/ign_ps3_cleaned.csv',
    'data/cleaned/ign_ps2_cleaned.csv',
    'data/cleaned/ign_ps1_cleaned.csv',
    'data/cleaned/ign_pc_cleaned.csv',
    'data/cleaned/ign_n64_cleaned.csv',
    'data/cleaned/ign_ngage_cleaned.csv',
    'data/cleaned/ign_mac_cleaned.csv',
    'data/cleaned/ign_iphone_cleaned.csv',
    'data/cleaned/ign_ipad_cleaned.csv',
    'data/cleaned/ign_gba_cleaned.csv',
    'data/cleaned/ign_gb_cleaned.csv',
    'data/cleaned/ign_gamecube_cleaned.csv',
    'data/cleaned/ign_ds_cleaned.csv',
    'data/cleaned/ign_dreamcast_cleaned.csv',
    'data/cleaned/ign_android_cleaned.csv',
    'data/cleaned/ign_3ds_cleaned.csv',
    'data/cleaned/ign_xboxx_cleaned.csv',
    'data/cleaned/ign_xboxone_cleaned.csv',
    'data/cleaned/ign_xbox360_cleaned.csv',
    'data/cleaned/ign_xbox_cleaned.csv',
    'data/cleaned/ign_wiiu_cleaned.csv',
    'data/cleaned/ign_wii_cleaned.csv',
    'data/cleaned/ign_vita_cleaned.csv',
    'data/cleaned/ign_switch_cleaned.csv',
    'data/cleaned/ign_psp_cleaned.csv',
    'data/cleaned/ign_ps5_cleaned.csv'
]

# Create an empty list to store the DataFrames
data_frames = []

# Loop through the file paths
for file_path in file_paths:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    # Append the DataFrame to the list
    data_frames.append(df)

# Concatenate all the DataFrames in the list
df_merged = pd.concat(data_frames)
```
